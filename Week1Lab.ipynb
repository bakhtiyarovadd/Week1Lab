# Problem 1


```
scores = [72, 85, 90,68,88]

print("Scores", scores)
average = sum(scores)/len(scores)
print("Average", average)

high_score = max(scores)
print("High Score", high_score)
lowes_score = min(scores)
print("Lowes Score", lowes_score)
```
