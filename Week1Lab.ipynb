Problem 3


```
scores = [45 , 67, 82, 59]
def pass_fail(score):
  if score >= 60:
    return "Pass"
  else:
    return "Fail"

for score in scores:
    print("Score", score, "- >", pass_fail(score))
```

```
sizes = [500 , 800 , 1000 , 1200]
prices = [50 , 80, 100 , 130]
for i in range(len(sizes)):
  print("Size", sizes[i], "sq.m- > Price", prices[i])
is_increasing = True
for i in range(1, len(prices)):
  if prices[i] < prices[i-1]:
    is_increasing = False
if is_increasing:
  print("Positive trend")
else:
  print("Negative trand")
```

Problem 6


```
scores = [35, 58, 62, 74, 89, 45]

for i in range(len(scores)):
    if scores[i] < 50:
        category = "Low"
    elif scores[i] <= 70:
        category = "Medium"
    else:
        category = "High"

    print("Score", scores[i], "->", category)
```

```
ages = [18, 20, 22, 24, 26]
heights = [160, 165, 170, 175, 180]

for i in range(len(ages)):
    print("Age", ages[i], "-> Height", heights[i])

sum_age = 0
for i in range(len(ages)):
    sum_age += ages[i]
avg_age = sum_age / len(ages)
print("Average age:", avg_age)

sum_height = 0
for i in range(len(heights)):
    sum_height += heights[i]
avg_height = sum_height / len(heights)
print("Average height:", avg_height)

is_increasing = True
for i in range(1, len(heights)):
    if heights[i] < heights[i-1]:
        is_increasing = False

if is_increasing:
    print("Height increases with age")
else:
    print("Height does not increase with age")

```
